```python
import pandas as pd
df = pd.read_csv('Mentalhealth.csv')
```

```python
df
```

```python
df = df.drop("topic", axis = 1)
```

```python
df
```

```python
lines = ['Readme', 'How to write text files in Python']
with open('readme.txt', 'w') as f:
    for line in lines:
        f.write(line)
        f.write('\n')
```

```python
problems = df["Problems"].tolist()
answers = df["Answers"].tolist()
```

```python
with open('mentalhealth.txt', 'w', encoding='utf-8') as f:
    for i in range(len(problems)):
        f.write(f"problem: {problems[i]}")
        f.write('\n')
        f.write(f"answer: {answers[i]}")
        f.write('\n')
```
